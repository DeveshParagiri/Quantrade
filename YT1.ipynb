```python
aapl = close.AAPL.to_frame()
```

```python
aapl["lag1"] = aapl.shift(periods=1)
```

```python
aapl["diff"] = aapl.AAPL.sub(aapl.lag1)
```

```python
aapl["% change"] = aapl.AAPL.div(aapl.lag1).sub(1).mul(100)

#OR aapl.AAPL.pct_change(periods=1).mul(100)
```

```python
aapl.rename(columns = {'% change': 'change'}, inplace=True)
```

```python
aapl
```

```python
del aapl["lag1"]
```

```python
del aapl["diff"]
```

```python
aapl
```

```python
aapl.AAPL.resample("BM").last().pct_change(periods=1).mul(100)
```

```python
del aapl["change"]
```

```python
ret = aapl.pct_change().dropna()
ret
```

```python
ret.plot(kind="hist", figsize=(12,8), bins=100)
plt.show()
```

```python
daily_mean_ret = ret.mean()
```

```python
var_daily = ret.var()
var_daily
```

```python
std_daily = np.sqrt(var_daily)
std_daily
```

```python
ret.std()
```

```python
annual_mean_ret = daily_mean_ret*252
annual_mean_ret
```

```python
annual_var_ret = var_daily*252
annual_var_ret
```

```python
annual_std_ret = np.sqrt(annual_var_ret)
annual_std_ret
```

```python
ticker = ['SPY','AAPL','KO','IBM','DIS','MSFT']
```

```python
stocks = yf.download(ticker, start="2010-01-01", end="2021-01-01")
```

```python
stocks
```

```python
close = stocks.loc[:,"Close"].copy()
```

```python
close
```

```python
normclose = close.div(close.iloc[0]).mul(100)

normclose.plot(figsize=(15,8),fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
ret=close.pct_change().dropna()
```

```python
ret.head()
```

```python
summary = ret.describe().T.loc[:,["mean","std"]]
summary
```

```python
summary["mean"]=summary["mean"]*252
summary["std"]=summary["std"]*np.sqrt(252)
summary
```

```python
summary.plot.scatter(x="std",y="mean",figsize=(12,8), s=50, fontsize=10)

for i in summary.index:
    plt.annotate(i,xy=(summary.loc[i,"std"]+0.002, summary.loc[i,"mean"]+0.002),size=10)

plt.xlabel("Annual risk(std)", fontsize=15)
plt.ylabel("Annual return", fontsize=15)
plt.title("Risk/Return", fontsize=25)
plt.show()
```
